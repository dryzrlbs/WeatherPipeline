```python
import requests
import json

# Open-Meteo API URL'si (Örnek URL, parametrelerini kendi ihtiyaçlarına göre ayarlayabilirsin)
url = "https://archive-api.open-meteo.com/v1/archive?start_date=2025-04-04&end_date=2025-04-05&latitude=60.1695&longitude=24.9354&hourly=temperature_2m,rain,snowfall,weather_code,soil_temperature_0_to_7cm,relative_humidity_2m,precipitation,cloudcover,wind_speed_10m,wind_direction_10m,pressure_msl&timezone=Europe/Helsinki"

# API'ye GET isteği gönder
response = requests.get(url)


```

```python
# Eğer istek başarılı olduysa (HTTP Status 200)
if response.status_code == 200:
    # JSON yanıtını al
    api_data = response.json()

    # Dönen veriyi ekrana yazdır
    print(json.dumps(api_data, indent=2))  # JSON formatında, okunabilir şekilde yazdır
else:
    print(f"API isteği başarısız oldu. Status Code: {response.status_code}")
```
